```python
rng = np.random.default_rng(100500)
L = rng.random(100)
sum(L)
```

```python
mean = [0, 0]
cov = [
    [1, 2],
    [2, 5]
]

X = rng.multivariate_normal(mean, cov, 100)
X.shape
```

```python
import matplotlib.pyplot as plt

# plt.style.use('seaborn-whitegrid')
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
includes = np.random.choice(X.shape[0], 10, replace=False)
includes
```

```python
selection = X[includes]
selection
```

```python
plt.scatter(X[:, 0], X[:, 1], alpha=0.3)
plt.scatter(selection[:, 0], selection[:, 1], facecolor='none', edgecolor='black', s=200)
plt.show()
```
